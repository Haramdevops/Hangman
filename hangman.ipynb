```python
import random

# Hangman ASCII Art Function
def draw_hangman(stage):
    stages = [
        """
           ------
           |    |
           |    
           |   
           |  
           |
        """,
        """
           ------
           |    |
           |    O
           |   
           |  
           |
        """,
        """
           ------
           |    |
           |    O
           |    |
           |  
           |
        """,
        """
           ------
           |    |
           |    O
           |   /|
           |  
           |
        """,
        """
           ------
           |    |
           |    O
           |   /|\\
           |  
           |
        """,
        """
           ------
           |    |
           |    O
           |   /|\\
           |   /
           |
        """,
        """
           ------
           |    |
           |    O
           |   /|\\
           |   / \\
           |
        """
    ]
    print(stages[stage])

# Word list for Hangman
word_list = ["undertaker", "john", "harry", "smith"]
word1 = random.choice(word_list)
hint = word1[0]
print("Letter starts with:", hint)
life = 6  # Increased lives for better gameplay
print("Lives:", life)

# Placeholders for guessed letters
placeholders = ["_"] * len(word1)
print(" ".join(placeholders))

while life > 0:
    draw_hangman(6 - life)  # Show Hangman progress

    guess = input("\nGuess a letter: ").lower()

    if guess in word1:
        print("Good Guess!")
        for index, letter in enumerate(word1):
            if letter == guess:
                placeholders[index] = guess
    else:
        life -= 1
        print(f"Wrong guess! Lives remaining: {life}")

    print(" ".join(placeholders))

    if "_" not in placeholders:
        print(f"\nðŸŽ‰ Congratulations! You guessed the word: {word1}")
        break

    if life == 0:
        draw_hangman(6)  # Show full Hangman before game over
        print(f"\nðŸ’€ Game Over! The word was: {word1}")
        break
```
